```python
import spacy
import pandas as pd
import numpy as np
import pickle

nlp = spacy.load('en_core_web_lg')
```

```python
info = pd.read_csv('InstitutionCampus.csv')
```

```python
info.head()
```

```python
universities = info['LocationName'].to_frame()
universities.head()
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
universities['vectors'] = universities.loc[:, 'LocationName'].apply(lambda x: nlp(x.lower()).vector)
```

```python
universities.head()
```

```python
import pickle
```

```python
with open('universities_vectors','wb') as f: pickle.dump(universities, f)
```
